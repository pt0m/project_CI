```python
from time import time
import cv2
import argparse
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pdb
from sklearn.feature_extraction.image import extract_patches_2d
from sklearn.feature_extraction.image import reconstruct_from_patches_2d
from sklearn.linear_model import OrthogonalMatchingPursuit
from sklearn.datasets import make_sparse_coded_signal
from sklearn.decomposition import MiniBatchDictionaryLearning
from matplotlib import pyplot as plt
from skimage.exposure import rescale_intensity
from skimage.metrics import peak_signal_noise_ratio
```

```python
#Defining parameters
n_comp=100 #nbr of components in the dictionnary
n_iter=500 #nbr of iteration to learn the dictionnary
sigma=0.1 #std of noise added on original image
non_zero_coeff=5
image=plt.imread("samoyed-data/samoyed-1.jpg")
```

```python
plt.imshow(image)
```

```python
#For one image
def get_noisy_patches(image,sigma, dict_learning=False, channel=None):
    """Return the patches of the the noisy image
    In:
    :image : considered image
    :sigma (float): standard deviation of the noise added on the image
    :dict_learning(bool): if True, downsample image, to train dictionnary on downsampled noisy image
    :channel(bool): indicates the presence of several channels
    Out:
    :data: noisy patches
    :distorted: noisy image
    :mean: mean patch
    :std:  std patch
    """

    image = image / 255
    print("dict learning",dict_learning)
    print("TEST")
    #downsample for higher speed
    if dict_learning==True :
        image = image[::2, ::2] + image[1::2, ::2] + image[::2, 1::2] + image[1::2, 1::2]
        image /= 4.0
    
    print('Distorting image...')
    noisy_image= image.copy()

    if channel :
        height, width, channel = image.shape
        noisy_image += sigma * np.random.randn(height, width, channel)
    else:
        height, width = image.shape
        noisy_image += sigma * np.random.randn(height, width)
        cv2.imwrite('noisy.jpg', (noisy_image*255))
    print(noisy_image.shape)

    print('Extracting reference patches...')
    t0 = time()
    patch_size = (7, 7)
    #Reshaping distorted image in a collection of patches, collected in dedicated array
    patches = extract_patches_2d(noisy_image, patch_size)
    patches = patches.reshape(patches.shape[0], -1)
    mean = np.mean(patches, axis=0)
    std = np.std(patches, axis=0)
    patches -= mean
    patches /= std
    print('done in %.2fs.' % (time() - t0)) 
    
    return (patches, noisy_image,mean, std)

```

```python
#create database of images

images=[plt.imread("samoyed-data/samoyed-"+str(i)+".jpg") for i in range(1,5)]
len(images)
```

```python
#For multiple images
def get_noisy_patches(images,sigma, dict_learning=False, channel=None):
    """Return the patches of the the noisy image
    In:
    :images (list): database of images
    :sigma (float): standard deviation of the noise added on the image
    :dict_learning(bool): if True, downsample image, to train dictionnary on downsampled noisy image
    :channel(bool): indicates the presence of several channels
    Out:
    :data: noisy patches
    :distorted: noisy image
    :mean: mean patch
    :std:  std patch
    """
    patches=[]
    for image in images:
        print('image i',image.shape)
        image = image / 255
        #downsample for higher speed
        if dict_learning==True :
            image = image[::2, ::2] + image[1::2, ::2] + image[::2, 1::2] + image[1::2, 1::2]
            image /= 4.0
        noisy_image= image.copy()
        if channel :
            height, width, channel = image.shape
            noisy_image += sigma * np.random.randn(height, width, channel)
        else:
            height, width = image.shape
            noisy_image += sigma * np.random.randn(height, width)
            cv2.imwrite('noisy.jpg', (noisy_image*255))
        print(noisy_image.shape)

        print('Extracting reference patches...')
        t0 = time()
        patch_size = (7, 7)
        #Reshaping distorted image in a collection of patches, collected in dedicated array
        patches_im = extract_patches_2d(noisy_image, patch_size)
        patches_im = patches_im.reshape(patches_im.shape[0], -1)
        mean = np.mean(patches_im, axis=0)
        std = np.std(patches_im, axis=0)
        patches_im -= mean
        patches_im /= std
        print('done in %.2fs.' % (time() - t0))
        patches.append(patches_im)
    
    return (patches, noisy_image,mean, std)
```

```python
get_noisy_patches(images,sigma,dict_learning=True,channel=None)
```

```python
def ksvd(noisy_patches):
    print('Updating Dictionary')
    t0 = time()
    #Mini-batch dictionnary Learning
    dico = MiniBatchDictionaryLearning(n_components=n_comp, 
                                        alpha=2, 
                                        n_iter=n_iter)
    print('done in %.2fs.' % (time() - t0))
    V = dico.fit(noisy_patches).components_
    return V, dico
```

```python
channel=None
if len(image.shape) >2:
    channel = image.shape[2]
n0_data,noisy_image, mean, std= get_noisy_patches(image,sigma,channel=channel)
noisy_patches,noisy_image_train,_,_ = get_noisy_patches(noisy_image,sigma, dict_learning=True, channel=channel)
dict_final, dico = ksvd(noisy_patches)
dico.set_params(transform_algorithm='omp',transform_n_nonzero_coefs = non_zero_coeff)
code = dico.transform(n0_data) #Encode the data as a sparse combination of the dictionary atoms
denoised_patches = np.dot(code,dict_final)
denoised_patches*= std
denoised_patches += mean
#Reshaping the patches in the correct shape
denoised_patches = (denoised_patches.reshape(n0_data.shape[0], 7, 7, channel))
print('Reconstructing...')
reconstruction = reconstruct_from_patches_2d(denoised_patches, (image.shape[0], image.shape[1], channel))
reconstruction*=255
difference = image - reconstruction
error = np.sqrt(np.sum(difference ** 2))
print('Difference (norm: %.2f)' %error)
print('Finished reconstruction..')


#Plotting results
plt.figure(figsize=(20,10))
plt.subplot(131)
plt.imshow(image)
plt.title("original image")
plt.subplot(132)
plt.imshow(noisy_image)
plt.title("noisy image")
plt.subplot(133)
plt.imshow(reconstruction.astype('uint8'))
plt.title("reconstructed image")

#Comparing psnr score
psnr_noisy=peak_signal_noise_ratio(image,noisy_image)
print("noisy psnr: ",psnr_noisy,"\n")
psnr_denoised=peak_signal_noise_ratio(image,reconstruction.astype('uint8'))
print("denoised psnr",psnr_denoised)
```
